```python
#data taking
data=pd.read_csv('Data_for_UCI_named.csv')
print(data.head())
y=data['stabf'].values
print("Target: ",y)
X=data.drop(['stabf'],axis=1)
y=np.where(y=='stable',1,0)

print("\nInput:\n",X)
print("Target: ",y)
```

```python
#Splitting data
#we have methods to split data some of them are 
#1. TRain test split
#2. sampling : we can split data using sampling method from numpy libray rather than sklearn

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
##developing model
w=np.empty((13,2),dtype=np.int32)
w[:,0]=np.random.uniform(low=0,high=2,size=(13,))
w[:,1]=np.random.uniform(low=0,high=2,size=(13,))
lrate= 0.5
itr=2
D=[0,0]
correct=0
C=0
print('learning rate of this epoch is',lrate)
for _ in range(itr):
    print('Epoch is',_+1)
    for i in range(1000):
        C=C+1
        for j in range(w.shape[1]):
             temp=0
             for k in range(w.shape[0]):
                temp = temp + ((w[k,j]-X_train.iloc[i,k])**2)
             D[j]=temp
        if(D[0]<D[1]):
            J=0
        else:
            J=1
        print('winning unit is',J)
        print('weight updation ...')
        print('current target',y_train[i])
        if J==y_train[i]:
            correct=correct+1
            for m in range(13):
                w[m,J]=w[m,J] + (lrate *(X_train.iloc[i,m]-w[m,J]))
        else:
            for m in range(13):
                w[m,J]=w[m,J] - (lrate *(X_train.iloc[i,m]-w[m,J]))
        print('Updated weights\n',w)
    print("total target used",C)
    accuracy=(correct/C)*100
    print("TARGET MATCHING WITH WINNING UNIT(accuracy)",accuracy)    
    C=0
    correct=0
    lrate=lrate*(1-((_+1)/2))
    print(' updated learning rate after ',_+1,' epoch is',lrate)
```
