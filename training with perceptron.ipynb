```python
#data taking
data=pd.read_csv('Data_for_UCI_named.csv')
print(data.head())
y=data['stabf'].values
print("Target: ",y)
X=data.drop(['stabf'],axis=1)
y=np.where(y=='stable',1,-1)
print("\nInput:\n",X)
print("Target: ",y)
```

```python
data.groupby('stabf').size()
```

```python
 
#scaling data to some set for better training the model    
from sklearn.preprocessing import StandardScaler
 scaler=StandardScaler().fit(X)
 rescaledX=scaler.transform(X)
 print(rescaledX.shape)
 rescaledX[0:5,:]
```

```python
#Scaled values
rescaledX
```

```python
#Splitting data
#we have methods to split data some of them are 
#1. TRain test split
#2. sampling : we can split data using sampling method from numpy libray rather than sklearn

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
        rescaledX,y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Developing the model
#Perceptron model
class Perceptron(object):
        def __init__(self, eta, n_iter):
                #fprint(X)
                self.eta = eta
                self.n_iter = n_iter
        def fit(self, X, y):
                self.w_ = np.zeros(1 + X.shape[1])
                self.errors_=[]
                plot_X=[]
                plot_y=[]
                for _ in range(self.n_iter):
                        errors=0
                        flag=0
                        for xi, target in zip(X, y):
                                plot_X.append(xi)
                                plot_y.append(self.predict(xi))
                                error = target - self.predict(xi)
                                if error != 0:
                                        flag=1
                                        update = self.eta * error
                                        self.w_[1:] += update * xi
                                        self.w_[0] += update
                                        errors+=int(update!=0.0)
                        self.errors_.append(errors)
                        if(not flag):
                            break
                return self
        def net_input(self, X):
                return np.dot(X, self.w_[1:]) + self.w_[0]
        def predict(self, X):
                return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
#fitting the data intothe perceptron model
""""from sklearn.linear_model import Perceptron"""
ppn=Perceptron(n_iter=400, eta=0.4)
#print(ppn)
print(X_train)
ppn.fit(X_train, y_train)
```

```python
#testing the data
y_pred=ppn.predict(X_test)
```

```python
#printing the accuracy of testing and misclassified sample
ms=(y_test!=y_pred).sum()
print('misclassified samples: %d'%(y_test!=y_pred).sum())#compute
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print('Predicting Accuracy:%.2f'%accuracy_score(y_test,y_pred))
```

```python
#pickling the trained model into the file
import pickle
with open("perceptron_trained_model","wb") as f:
    pickle.dump(ppn,f)
```

```python
#plotting the graph for errors occured while training 
plt.plot(range(1,len(ppn.errors_)+1),ppn.errors_,marker='o')
plt.xlabel("Number of iterations")
plt.ylabel("ERROR VALUES")
plt.show()
```
